# 3. High correlation between two columns

what is High correlation?
> High correlation between two variables means they have similar trends and are likely to carry similar information.

How it's work.
1. Basically,calculate the correlation between independent numerical variables that are numerical in nature.
2. Then If the correlation coefficient crosses a certain threshold value, we can drop one of the variables.

```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('travel-times.csv')
df
```

```python
# Check correlation  beteeen columns
cor_matrix = df.corr().abs()
cor_matrix
```

```python
# corresponding correlation
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

```python
# check correlation in variable  and if morethen 0.9% then romove ehos variable
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)]
print(to_drop)
```
